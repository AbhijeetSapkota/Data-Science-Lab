```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("./Health_heart_experimental.csv")
df.head(5)
df_new = df.groupby('age').sum(numeric_only=True).reset_index()
```

### Line Chart

A line chart, also known as a line graph or line plot, is a chart that uses line segments to connect data points and show changes in value.

```python
sns.lineplot(x=df['age'],y=df['BMI'])
```

The above lineplot shows age and BMI of the data. We can see BMI with respect to age in the above Graph.

----

### Pie Chart

A pie chart is a way of representing data in a circular graph.

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

The above pie chart is of sex.We can see inthe above pie chart that There is around equal number of male and female in the data taken.

```python
df['indication'].value_counts().plot(kind='pie',autopct='%0.2f')
```

We can see that more people are healthy then unhealthy.

---

### Bar Plot

A bar plot uses rectangular bars to represent data categories, with bar length or height proportional to their values.

```python
sns.countplot(data=df, x='sex',hue ='indication')
```

We can see that there are more healthy people in both sex.

---

### Box Plot

A box plot uses boxes and lines to depict the distributions of one or more groups of numeric data. Box limits indicate the range of the central 50% of the data, with a central line marking the median value.

```python
sns.boxplot(df['weightKg'])
```

The above box plot is of weight. We can see in the above box plot that,there is are no outliers in the data taken. The median is around 70kg. Also we can see that 50% of people (Q1 -Q3) are of weight around 55-85kg.

```python
sns.boxplot(x='sex',y='age',hue = 'indication',data=df)
```

---

### Histogram

A histogram is a chart that plots the distribution of a numeric variable's values as a series of bars. Each bar typically covers a range of numeric values called a bin or class.

```python
plt.hist(df['heightCm'])
```

We can see that this is a left-skewed histogram. People with lower height are less while higher height are high.

```python
plt.hist(df['weightKg'])
```

We can see that this is a right-skewed histogram. People with lower weight are high while higer weight are low.

```python
plt.hist(df['age'])
```

We can see that this is a symmetric histogram as people across all age are equal in count.

---

### Scatter Plot

 A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point.

```python
sns.scatterplot(x='heightCm',y='weightKg',hue='indication' ,data=df)
plt.show()
```

This shows that as height increases, weight tends to increase as well but only those with a balanced proportion of height and weight are considered to be in optimal health.
